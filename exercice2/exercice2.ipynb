## Exercice 2 - Question 0

Considérons la fonction

$$
f(x)=x^3.
$$

Sa dérivée est

$$
f'(x)=3x^2.
$$

Par conséquent,

$$
f'(0)=0.
$$

Cependant, $x=0$ n'est ni un minimum local ni un maximum local.

En effet, pour tout $\varepsilon>0$, il existe des points $x\in(-\varepsilon,0)$ tels que

$$
f(x)<f(0),
$$

et des points $x\in(0,\varepsilon)$ tels que

$$
f(x)>f(0).
$$

Ainsi, dans tout voisinage de $0$, la fonction prend à la fois des valeurs plus petites et plus grandes que $f(0)$.

Par conséquent, $f(0)$ n'est pas un extremum local, bien que

$$
f'(0)=0.
$$

## Exercice 2 - Question 1

Considérons le cadre suivant :

$$
X=\{0\}
$$

et

$$
Y=
\begin{cases}
0 & \text{avec probabilité } 0.75,\\
4 & \text{avec probabilité } 0.25.
\end{cases}
$$

Comme $X$ est constant, tout prédicteur est une constante.

### Bayes predictor pour la perte quadratique

On sait que

$$
f^*_{\mathrm{squared}}
=
\mathbb E[Y].
$$

Or

$$
\mathbb E[Y]
=
0\times0.75
+
4\times0.25
=
1.
$$

Ainsi,

$$
f^*_{\mathrm{squared}}=1.
$$

### Bayes predictor pour la perte absolue

Le prédicteur de Bayes associé à la perte absolue est la médiane.

Comme

$$
P(Y\le 0)=0.75>0.5,
$$

la médiane vaut

$$
f^*_{\mathrm{absolute}}=0.
$$

On obtient donc

$$
f^*_{\mathrm{absolute}}
\neq
f^*_{\mathrm{squared}}.
$$

```python
import numpy as np

rng = np.random.default_rng(42)

n_test = 1_000_000

Y = rng.choice([0, 4], size=n_test, p=[0.75, 0.25])

f_squared = 1.0
f_absolute = 0.0

risk_squared_estimator = np.mean(np.abs(Y - f_squared))
risk_absolute_estimator = np.mean(np.abs(Y - f_absolute))

print("Absolute loss of f_squared* :", risk_squared_estimator)
print("Absolute loss of f_absolute* :", risk_absolute_estimator)
```

## Exercice 2 - Question 2

Pour un $x\in X$ fixé, on considère

$$
g(z)
=
\mathbb E\left[|Y-z|\mid X=x\right]
=
\int_{-\infty}^{+\infty}
|y-z|\,p_{Y|X=x}(y)\,dy.
$$

On écrit

$$
g(z)
=
\int_{-\infty}^{z}(z-y)p(y)\,dy
+
\int_{z}^{+\infty}(y-z)p(y)\,dy,
$$

où

$$
p(y)=p_{Y|X=x}(y).
$$

En dérivant :

$$
g'(z)
=
\int_{-\infty}^{z}p(y)\,dy
-
\int_{z}^{+\infty}p(y)\,dy.
$$

Comme

$$
\int_{-\infty}^{+\infty}p(y)\,dy=1,
$$

on obtient

$$
g'(z)
=
2F_{Y|X=x}(z)-1,
$$

où $F_{Y|X=x}$ est la fonction de répartition conditionnelle.

Le minimum est atteint lorsque

$$
g'(z)=0.
$$

Ainsi,

$$
F_{Y|X=x}(z)=\frac12.
$$

Par définition, une telle valeur de $z$ est une médiane de la loi conditionnelle
$Y\mid X=x$.

On conclut que le prédicteur de Bayes associé à la perte absolue est

$$
f^*_{l_{\mathrm{absolute}}}(x)
=
\operatorname{Med}(Y\mid X=x).
$$
