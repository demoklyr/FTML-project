## Context

OLS framework:
- Input space $\mathcal{X} = \mathbb{R}^d$, output space $\mathcal{Y} = \mathbb{R}$
- Loss function: squared loss $\ell(y, y') = (y - y')^2$
- Hypothesis space: $\mathcal{F} = \{x \mapsto x^\top \theta,\ \theta \in \mathbb{R}^d\}$
- Design matrix $X \in \mathbb{R}^{n \times d}$, assumed injective (so $d \leq n$)

The OLS estimator minimising empirical risk $R_n(\theta) = \frac{1}{n}\|y - X\theta\|^2$:
$$\hat{\theta} = (X^\top X)^{-1} X^\top y$$

Linear model (fixed design):
$$y = X\theta^* + \varepsilon, \qquad \varepsilon \sim \mathcal{N}(0, \sigma^2 I_n)$$

<u>Goal:</u> prove Proposition 1:
$$\mathbb{E}\!\left[R_X(\hat{\theta})\right] = \frac{n-d}{n}\,\sigma^2$$

## Question 1

### Bayes risk

In the linear model $y = X\theta^* + \varepsilon$ with $\varepsilon \sim \mathcal{N}(0, \sigma^2 I_n)$, the conditional distribution of $y_i$ given $x_i$ is $\mathcal{N}(x_i^\top \theta^*, \sigma^2)$. The Bayes predictor for the squared loss is the conditional expectation $f^*(x) = x^\top \theta^*$, and the **Bayes risk** is:
$$R^* = \mathbb{E}\!\left[(y - x^\top \theta^*)^2\right] = \mathbb{E}[\varepsilon^2] = \sigma^2$$

### Comparison

Proposition 1 states:
$$\mathbb{E}\!\left[R_X(\hat{\theta})\right] = \frac{n-d}{n}\,\sigma^2$$

Since $n \geq d \geq 1$, we have $\frac{n-d}{n} \in [0,1]$, hence:
$$\frac{n-d}{n}\sigma^2 \leq \sigma^2 \implies \mathbb{E}\!\left[R_X(\hat{\theta})\right] \leq R^*$$

**The expected empirical risk of $\hat{\theta}$ is smaller than the Bayes risk.**

This is not a contradiction: the Bayes risk is a generalisation error (over new unseen points), whereas $R_X(\hat{\theta})$ is the training risk on the same data used to fit $\hat{\theta}$. The OLS estimator is chosen to minimise the empirical risk, so it naturally achieves a lower training error than the true noise level.

### Behaviour as a function of $n$ and $d$

We can rewrite:
$$\mathbb{E}\!\left[R_X(\hat{\theta})\right] = \left(1 - \frac{d}{n}\right)\sigma^2$$

The ratio $\frac{d}{n}$ measures the complexity of the model relative to the number of observations.

- When $d \ll n$: $\frac{d}{n} \approx 0$, so $\mathbb{E}[R_X(\hat{\theta})] \approx \sigma^2 = R^*$. The model does not have enough degrees of freedom to memorise the noise specific to the training data. The empirical risk is close to the irreducible noise level $\sigma^2$, and is a reliable indicator of the generalisation error.

- When $d \to n$: $\frac{d}{n} \to 1$, so $\mathbb{E}[R_X(\hat{\theta})] \to 0$. The model has enough degrees of freedom to fit the training data almost exactly, absorbing the noise $\varepsilon$ into $\hat{\theta}$. The training error vanishes, but $\hat{\theta}$ is far from $\theta^*$. The model **overfits**.

So, a small value of $\mathbb{E}[R_X(\hat{\theta})]$ is not a sign of a good estimator, it is a sign that the model has memorised the training data.

## Question 2

We want to show that:
$$\mathbb{E}\!\left[R_n(\hat{\theta})\right] = \mathbb{E}_\varepsilon\!\left[\frac{1}{n}\left\|(I_n - X(X^\top X)^{-1}X^\top)\,\varepsilon\right\|^2\right]$$

By definition the OLS estimator is $\hat{\theta} = (X^\top X)^{-1}X^\top y$

So we have:
$$y - X\hat{\theta} = y - X(X^\top X)^{-1}X^\top y = (I_n - X(X^\top X)^{-1}X^\top)\,y$$

Replacing y by $X\theta^* + \varepsilon$:

$$y - X\hat{\theta} = (I_n - X(X^\top X)^{-1}X^\top)(X\theta^* + \varepsilon)$$

The term in $X\theta^*$ vanishes:
$$(I_n - X(X^\top X)^{-1}X^\top)X\theta^* = X\theta^* - X\underbrace{(X^\top X)^{-1}X^\top X}_{= I_d}\theta^* = X\theta^* - X\theta^* = 0$$

Hence:
$$y - X\hat{\theta} = (I_n - X(X^\top X)^{-1}X^\top)\,\varepsilon$$

Substituting into $R_n(\hat{\theta}) = \frac{1}{n}\|y - X\hat{\theta}\|^2$:
$$R_n(\hat{\theta}) = \frac{1}{n}\left\|(I_n - X(X^\top X)^{-1}X^\top)\,\varepsilon\right\|^2$$

In the fixed design setting, $X$ is deterministic and the only source of randomness is $\varepsilon$. Taking the expectation over $\varepsilon$:
$$\mathbb{E}\!\left[R_n(\hat{\theta})\right] = \mathbb{E}_\varepsilon\!\left[\frac{1}{n}\left\|(I_n - X(X^\top X)^{-1}X^\top)\,\varepsilon\right\|^2\right]$$

## Question 3

Let $A \in \mathbb{R}^{n, n}$ 

We want to show that:
$$\sum_{(i,j) \in [1,n]^2} A_{ij}^2 = \mathrm{tr}(A^\top A)$$

By definition of matrix multiplication, the $(i,i)$ entry of $A^\top A$ is:
$$(A^\top A)_{ii} = \sum_{k=1}^{n} (A^\top)_{ik}\, A_{ki} = \sum_{k=1}^{n} A_{ki}\, A_{ki} = \sum_{k=1}^{n} A_{ki}^2$$

Summing over the diagonal:
$$\mathrm{tr}(A^\top A) = \sum_{i=1}^{n} (A^\top A)_{ii} = \sum_{i=1}^{n} \sum_{k=1}^{n} A_{ki}^2 = \sum_{(i,j) \in [1,n]^2} A_{ij}^2$$

*(In the last step we simply renamed the summation index $k \to j$)*

## Question 4

We want to show that:
$$\mathbb{E}_\varepsilon\!\left[\frac{1}{n}\|A\varepsilon\|^2\right] = \frac{\sigma^2}{n}\,\mathrm{tr}(A^\top A)$$

We expand the squared norm:
$$\|A\varepsilon\|^2 = \sum_{i=1}^n (A\varepsilon)_i^2 = \sum_{i=1}^n \left(\sum_{j=1}^n A_{ij}\varepsilon_j\right)^2 = \sum_{i=1}^n \sum_{j=1}^n \sum_{k=1}^n A_{ij}A_{ik}\varepsilon_j\varepsilon_k$$

Taking expectations, and using the fact that $A$ is deterministic (fixed design), 
so all $A_{ij}$ are constants that can be taken outside the expectation:

$$\mathbb{E}[\|A\varepsilon\|^2] 
= \mathbb{E}\!\left[\sum_{i=1}^n \sum_{j=1}^n \sum_{k=1}^n A_{ij}A_{ik}\varepsilon_j\varepsilon_k\right]$$

By linearity of $\mathbb{E}$, and since $A_{ij}, A_{ik}$ are constants:
$$= \sum_{i=1}^n \sum_{j=1}^n \sum_{k=1}^n A_{ij}A_{ik}\,\mathbb{E}[\varepsilon_j\varepsilon_k]$$

Since the $\varepsilon_i$ are independent with mean $0$ and variance $\sigma^2$:
$$\mathbb{E}[\varepsilon_j \varepsilon_k] = \begin{cases} \sigma^2 & \text{if } j = k \\ 0 & \text{if } j \neq k \end{cases} = \sigma^2\,\mathbf{1}_{j=k}$$

Substituting:
$$= \sum_{i=1}^n \sum_{j=1}^n \sum_{k=1}^n A_{ij}A_{ik}\,\sigma^2\,\mathbf{1}_{j=k}$$

The indicator $\mathbf{1}_{j=k}$ kills all terms where $j \neq k$, leaving only $j = k$:
$$= \sigma^2 \sum_{i=1}^n \sum_{j=1}^n A_{ij}A_{ij} = \sigma^2 \sum_{i=1}^n \sum_{j=1}^n A_{ij}^2 = \sigma^2 \sum_{(i,j) \in [1,n]^2} A_{ij}^2$$

By the result of **Question 3**:
$$= \sigma^2\,\mathrm{tr}(A^\top A)$$

Dividing by $n$:
$$\mathbb{E}_\varepsilon\!\left[\frac{1}{n}\|A\varepsilon\|^2\right] = \frac{\sigma^2}{n}\,\mathrm{tr}(A^\top A)$$

## Question 5

Let $A = I_n - X(X^\top X)^{-1}X^\top$

We want to show that: $$A^\top A = A$$

### Step 1: Show that $A$ is symmetric ($A^\top = A$)

$$A^\top = (I_n - X(X^\top X)^{-1}X^\top)^\top = I_n^\top - (X(X^\top X)^{-1}X^\top)^\top$$

We expand the second term using $(PQR)^\top = R^\top Q^\top P^\top$:
$$(X(X^\top X)^{-1}X^\top)^\top = X\,(X^\top X)^{-\top}\,X^\top$$

Since $X^\top X$ is symmetric, $(X^\top X)^{-\top} = (X^\top X)^{-1}$, hence:
$$A^\top = I_n - X(X^\top X)^{-1}X^\top = A$$

$A$ is symmetric.

### Step 2: Show that $A^2 = A$

Since $A^\top = A$, showing $A^\top A = A$ is equivalent to showing $A^2 = A$.

$$A^2 = (I_n - X(X^\top X)^{-1}X^\top)^2$$
$$= I_n - 2X(X^\top X)^{-1}X^\top + X(X^\top X)^{-1}X^\top X(X^\top X)^{-1}X^\top$$

The last term simplifies using $X^\top X(X^\top X)^{-1} = I_d$:
$$X(X^\top X)^{-1}\underbrace{X^\top X(X^\top X)^{-1}}_{=\,I_d}X^\top = X(X^\top X)^{-1}X^\top$$

Hence:
$$A^2 = I_n - 2X(X^\top X)^{-1}X^\top + X(X^\top X)^{-1}X^\top = I_n - X(X^\top X)^{-1}X^\top = A$$

### Conclusion: 
$$A^\top A = A^2 = A$$

## Question 6

Combining the results of questions 2, 4 and 5 and with $A = I_n - X(X^\top X)^{-1}X^\top$:

- Question 2 gives us:
$$\mathbb{E}\!\left[R_n(\hat{\theta})\right] = \mathbb{E}_\varepsilon\!\left[\frac{1}{n}\|A\varepsilon\|^2\right]$$

- Question 4 gives us:
$$\mathbb{E}_\varepsilon\!\left[\frac{1}{n}\|A\varepsilon\|^2\right] = \frac{\sigma^2}{n}\,\mathrm{tr}(A^\top A)$$

- Question 5 gives us $A^\top A = A$, hence:
$$\frac{\sigma^2}{n}\,\mathrm{tr}(A^\top A) = \frac{\sigma^2}{n}\,\mathrm{tr}(A)$$

It remains to compute $\mathrm{tr}(A)$:
$$\mathrm{tr}(A) = \mathrm{tr}(I_n - X(X^\top X)^{-1}X^\top)$$
$$= \mathrm{tr}(I_n) - \mathrm{tr}(X(X^\top X)^{-1}X^\top)$$

By the cyclic property of the trace $\mathrm{tr}(PQR) = \mathrm{tr}(RPQ)$:
$$\mathrm{tr}(X(X^\top X)^{-1}X^\top) = \mathrm{tr}(X^\top X(X^\top X)^{-1}) = \mathrm{tr}(I_d) = d$$

Therefore:
$$\mathrm{tr}(A) = n - d$$

And finally:
$$\boxed{\mathbb{E}\!\left[R_n(\hat{\theta})\right] = \frac{\sigma^2}{n}(n - d) = \frac{n-d}{n}\,\sigma^2}$$

## Question 7

We want to find $\mathbb{E}\!\left[\frac{\|y - X\hat{\theta}\|^2}{n - d}\right]$

From Question 2 and the proof above:
$$\mathbb{E}\!\left[\|y - X\hat{\theta}\|^2\right] = \mathbb{E}\!\left[n \cdot R_n(\hat{\theta})\right] = n \cdot \frac{n-d}{n}\,\sigma^2 = (n-d)\,\sigma^2$$

Therefore:
$$\mathbb{E}\!\left[\frac{\|y - X\hat{\theta}\|^2}{n-d}\right] = \frac{(n-d)\,\sigma^2}{n-d} = \sigma^2$$

Which means it's an unbiased estimator of $\sigma^2$

## Question 8

### Simulation

From Proposition 1, we have $\mathbb{E}[R_n(\hat{\theta})] = \frac{n-d}{n}\sigma^2$, which gives us 
the following unbiased estimator of $\sigma^2$:

$$\hat{\sigma}^2 = \frac{n}{n-d}\,R_n(\hat{\theta}) = \frac{\|y - X\hat{\theta}\|^2}{n-d}$$

We verify this result numerically by simulating the linear model $y = X\theta^* + \varepsilon$ 
with known parameters $n$, $d$, $\sigma^2$ and $\theta^*$. For each simulation, we compute 
$\hat{\sigma}^2$ and average the results over a large number of runs. If the estimator is 
correct, the average should converge to the true value $\sigma^2$.

```python
import numpy as np

n, d = 100, 5
sigma2 = 7.5
theta_star = np.random.randn(d) # The true weights

# X is fixed
X = np.random.randn(n, d)

n_simulations = 10000
estimates = []

for _ in range(n_simulations):
    epsilon = np.random.randn(n) * np.sqrt(sigma2)
    y = X @ theta_star + epsilon
    
    theta_hat = np.linalg.inv(X.T @ X) @ X.T @ y
    residuals = y - X @ theta_hat
    
    estimate = np.dot(residuals, residuals) / (n - d) # formula from question 7
    estimates.append(estimate)

print(f"Estimated sigma^2: {np.mean(estimates):.4f}")
print(f"True value of sigma^2: {sigma2}")
```

### Observations

The simulation confirms that $\hat{\sigma}^2 = \frac{\|y - X\hat{\theta}\|^2}{n-d}$ is 
a reliable estimator of $\sigma^2$: the empirical average over the simulations converges 
to the true value $\sigma^2 = 7.5$.
